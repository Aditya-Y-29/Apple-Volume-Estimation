```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df_predicted = pd.read_csv('../Data/predicted_volume.csv')
X = df_predicted['predicted_volume'].tolist()
df_true = pd.read_csv('../Data/apple_volume_mapper.csv')
Y = df_true['Volume'].tolist()
# Removing element from 0th and 2nd index
Y = Y[1:2] + Y[3:]
```

```python
X_array = np.array(X).reshape(-1, 1)
Y_array = np.array(Y)

model = LinearRegression()
model.fit(X_array, Y_array)
```

```python
scaled_predicted_values = model.predict(X_array)
```

```python
mse = mean_squared_error(Y_array, scaled_predicted_values)
rmse =  mse ** 0.5
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
```

```python
percentage_errors = [abs((actual - predicted) / actual) * 100 for actual, predicted in zip(Y_array, scaled_predicted_values)]
mean_percenatge_errors = sum(percentage_errors) / len(percentage_errors)
```
